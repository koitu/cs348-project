```python
import pandas as pd
```

```python
duplicate_player_df = pd.read_csv('teams.csv', sep=',', header=0, dtype = str)[['team','name',]]
all_player_df = duplicate_player_df.drop_duplicates()

sample = all_player_df.sort_values(by=['team'])
```

```python
i = 100
output_string = ""
for index, row in sample.iterrows():
  i+=1
  output_string += "\t("+str(i)+", \""+row['name']+"\", \""+"\", \""+"\", \""+"\"),\n"
```

```python
print(output_string)
```

```python
j = 100
output_s = "{"
for index, row in sample.iterrows():
  i+=1
  output_s += "\""+row['name']+"\":"+str(i)+", "
output_s+="}"
print(output_s)
```
