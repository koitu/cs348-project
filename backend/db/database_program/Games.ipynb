```python
def inch_to_cm(height):
  foot = int(height[0])
  inch = int(height[2:-1])
  return(int((12*foot+inch)*2.58))
```

```python
for index, row in sample.iterrows():
  output_string += "\t("+row['playerId']+", '"+row['name']+"', '"+row['birthDate']+"', "+row['weight'][0:3]+", "+str(inch_to_cm(row['height']))+", '"+row['team']+"', NULL, "+str(2023-int(row['birthDate'][0:4]))+", '"+row['position']+"'),\n"
```
