```python
import pandas as pd
```

```python
all_player_df = pd.read_csv('allPlayersLookup.csv', sep=',', header=0, dtype = str)
all_player_df['team'] = all_player_df['team'].replace('L.A','LAK')
all_player_df['team'] = all_player_df['team'].replace('N.J','NJD')
all_player_df['team'] = all_player_df['team'].replace('S.J','SJS')
all_player_df['team'] = all_player_df['team'].replace('T.B','TBL')
# print(all_player_df[all_player_df.team=="ATL"].index)
all_player_df = all_player_df.drop(all_player_df[all_player_df.team=="ATL"].index)
# all_player_df = all_player_df.drop([ 184,  314,  554,  595,  938,  972, 1002, 1048, 1472, 1494, 1822,
#             1959, 2211, 2432, 2566, 2921])
all_player_df
```

```python
teamdf = all_player_df[['team']].drop_duplicates()
teamdf.sort_values(by=['team'])
```

```python
output_string = ""
```

```python
sample = all_player_df.loc[[5,200,400,800,1300,1600,1900,2300,2500,2900]]
sample
```

```python
def inch_to_cm(height):
  foot = int(height[0])
  inch = int(height[2:-1])
  return(int((12*foot+inch)*2.58))
```

```python
for index, row in sample.iterrows():
  output_string += "\t("+row['playerId']+", '"+row['name']+"', '"+row['birthDate']+"', "+row['weight'][0:3]+", "+str(inch_to_cm(row['height']))+", '"+row['team']+"', NULL, "+str(2023-int(row['birthDate'][0:4]))+", '"+row['position']+"'),\n"
```
